```python
class vect3d:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __add__(self, other):
        return vect3d(self.x + other.x, self.y + other.y, self.z + other.z)
    def __sub__(self, other):
        return vect3d(self.x - other.x, self.y - other.y, self.z - other.z)
    def __mul__(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    def __repr__(self):
        return f"({self.x},{self.y},{self.z})"
v1 = vect3d(1, 2, 3)
v2 = vect3d(9, 10, 19)
print(f"v1+v2 = {v1+v2}")
print(f"v1-v2 = {v1-v2}")
print(f"v1*v2 = {v1*v2}")
```

```python
class Positive:
    def __init__(self):
        self._name = None
    def __set_name__(self,name):
        self._name =name
    def __get__(self, instance,):
        return getattr(instance, self._name)
    def __set__(self, instance, value):
        if value < 0:
            raise ValueError(f"{self._name[1:]} must be non-negative")
        setattr(instance, self._name, value)
```

```python
class rectangle:
    def __init__(self,h,w):
        self.h=h
        self.w=w
    def area(self):
        return self.h *self.w
    def perim(self):
        return (self.h+self.w)*2
rec=rectangle(15,20)
print(rec.area())
print(rec.perim())   
```

```python
class employee:
    def __init__(self,name,id,sal):
        self.name=name
        self.id=id
        self.sal=sal
    @classmethod    
    def infor(cls,empstr):
        name,id,sal=map(str,empstr.split(","))
        return cls(name,id,sal)
    def display_employee_info(self):
        print(f"Name:{self.name}")
        print(f"Id :{self.id}")
        print(f"Sal: {self.sal}")
emp1=employee("omar","e560",10000)
emp1.display_employee_info()
emp2=employee.infor("omar,e580,150000")
emp2.display_employee_info()        
```

```python
class vehicle:
    def move(self):
        print("Vehicle is moving")
class bike(vehicle):
    def move(self):
        print("Bike is cycling")
class car(vehicle):
    def move(self):
        print("car is driving")    
bikeinst=bike()
carinst=car()
vec=[bikeinst,carinst]
for x in vec:
    x.move()                   
```

```python
import sqlite3
conn = sqlite3.connect("school.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INTEGER PRIMARY KEY,
    name TEXT,
    grade REAL
)
""")
students = [
    (1, "Omar", 90.5),
    (2, "Sara", 92),
    (3, "Ali", 85.5)
]
cursor.executemany("INSERT OR REPLACE INTO students VALUES (?, ?, ?)", students)
conn.commit()
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
import requests
from bs4 import BeautifulSoup
import csv
url = "http://books.toscrape.com"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
books = soup.find_all('article', class_='product_pod')
with open('simple_books.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['Book Title', 'Price'])
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        writer.writerow([title, price])
        print(f" title: {title} ----- price: {price}")
```
